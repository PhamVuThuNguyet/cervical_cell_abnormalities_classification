```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
loss = []
val_loss = []

accuracy = []
val_accuracy = []
```

```python
with open("./Epoch.txt", "r+") as f:
    i = 1
    for l in f.readlines():
        if i % 2 == 0:
            temp = l.split("-")
            print(temp[2:6])
            loss.append(float(temp[2].split(":")[1]))
            accuracy.append(float(temp[3].split(":")[1]))
            val_loss.append(float(temp[4].split(":")[1]))
            val_accuracy.append(float(temp[5].split(":")[1]))
        
        i += 1
```

```python
def draw(x, y, multi = False, label = None):
    fig, ax = plt.subplots()

    # add labels to axes
    ax.set_xlabel('Fold')
    ax.set_ylabel('Accuracy')
    
    plt.ylim([0, 1])
    
    if multi == True:
        for i, y_batch in enumerate(y):
            if label != None:
                plt.plot(x, y_batch, label=label[i])
            else:
                plt.plot(x, y_batch)
        plt.legend()
        
    else:
        plt.plot(x, y)
    
    plt.show()
```

```python
X = np.arange(1, 201)
X
```

```python
draw(x = X, y = [loss, val_loss], label = ["training", "validating"], multi = True)
```

```python
draw(x = X, y = [accuracy, val_accuracy], label = ["training", "validating"], multi = True)
```
